```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
back = np.array([11, 13, 15, 18, 21, 24, 28, 32, 36, 40])
front = np.array([24, 32, 42])

ratios = []

colors = ['r', 'g', 'b']

for i, front_teeth in enumerate(front, 0):
    for back_teeth in back:

        ratio = front_teeth / back_teeth
        plt.scatter(ratio, i*0.1, color=colors[i])

        if ratio not in ratios:
            ratios.append(ratio)

back = np.array([11, 13, 15, 18, 21, 24, 28, 32, 36, 40, 46, 52])
front = np.array([32])

for i, front_teeth in enumerate(front, 0):
    for back_teeth in back:

        ratio = front_teeth / back_teeth
        plt.scatter(ratio, 0.4, color='c')

        if ratio not in ratios:
            ratios.append(ratio)

plt.show()
```
