```python
import pybliometrics.scopus as sc
import torch as th
import pandas as pd
from tqdm import tqdm
```

```python
# sort for abstract

df = pd.read_csv('data/test_1.csv', sep=';')
new_df = pd.DataFrame()
```

```python
for i in tqdm(range(len(df.index))):

    row = df.iloc[i]
    desc = row['description']

    if type(desc) != type(None):
        new_df = pd.concat([new_df, row], axis=1)

new_df = new_df.T
new_df.to_csv('data/with_abs.csv', index=False)
```
